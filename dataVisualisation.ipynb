# Employment Rate Data Visualisation
The following program uses data on Employment Rate (aged 16 to 64), stored in CSV files retrieved from https://www.ons.gov.uk, to create easy to visualise and analyse graphics. It makes use of Pandas and Matplotlib Python Libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#read data file
data_file='series-051221.csv'
data = pd.read_csv(data_file)
```

```python
#show data in table
data
```

```python
#extract data and store into lists
year = list(data['Year'])
avg = list(data['AvgRate'])
q1 = list(data['Q1'])
q2 = list(data['Q2'])
q3 = list(data['Q3'])
q4 = list(data['Q4'])
jan = list(data['JAN'])
feb = list(data['FEB'])
mar = list(data['MAR'])
apr = list(data['APR'])
may = list(data['MAY'])
jun = list(data['JUN'])
jul = list(data['JUL'])
aug = list(data['AUG'])
sep = list(data['SEP'])
oct = list(data['OCT'])
nov = list(data['NOV'])
dec = list(data['DEC'])
```

```python
#scatter plot year-avgRate
plt.plot(year,avg)
plt.xlabel('Year')
plt.ylabel('Average Rate %')
plt.title('Employment Rate (aged 16 to 64) from 1971 to 2020')
plt.show()
```

```python
#quaters stacked bar graphic

#quarter 1
x1 = year
y1 = q1
plt.bar(x1, y1, label = 'Q1')

#quarter 2
x2 = year
y2 = q2
plt.bar(x2, y2, label = 'Q2')

#quarter 3
x3 = year
y3 = q3
plt.bar(x3, y3, label = 'Q3')

#quarter 4
x4 = year
y4 = q4
plt.bar(x4, y4, label = 'Q4')

#limit x and y values
plt.xlim(1970,2020)
plt.ylim(65,78)

#show legend in locked position
plt.legend(loc="upper left")

#label graphic
plt.xlabel('Years')
plt.ylabel('Employment Rate %')
plt.title('Employment Rate (aged 16 to 64) from 1971 to 2020 by quarters')

plt.show()
```

```python
#pie graph with month values for 1981
monthNames = ['January', 'February', 'March', 'April', 'May', 'June',
             'July', 'August', 'September', 'October', 'November', 'December']

monthValues = ['69.5','69.3', '69.1', '68.9', '68.7', '68.5', '68.4', 
               '68.2', '68.1', '67.9', '67.7', '67.6']
#design graph
plt.pie(monthValues, labels = monthNames,
        startangle=90, shadow = True,
        radius = 1.2, autopct = '%1.1f%%')

plt.title('Employment Rate (aged 16 to 64) during 1981',
         y=1.1)

plt.show()


```
